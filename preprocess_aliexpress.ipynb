```python
from sklearn.preprocessing import OrdinalEncoder

train_categorical_encoded = train_categorical.copy()
test_categorical_encoded = test_categorical.copy()

encoder = OrdinalEncoder()
train_categorical_encoded[train_categorical.columns] = encoder.fit_transform(train_categorical)
test_categorical_encoded[test_categorical.columns] = encoder.fit_transform(test_categorical)

train_categorical_encoded.head()
```

```python
train_click = ali_train['click']
train_conversion = ali_train['conversion']

test_click = ali_test['click']
test_conversion = ali_test['conversion']
```

```python
import json 

# load the validation indices
with open('./data/splits/aliexpress_val_indices.json', 'r') as f:
    cached_indices = json.load(f)

val_indices = np.array(cached_indices)
train_indices = np.setdiff1d(np.arange(len(ali_train)), val_indices)
```

```python
import h5py

data_root_aliexpress = f'{data_root}/AliExpress_NL/'

with h5py.File(data_root_aliexpress + 'train_val_test.h5', 'w') as f:
    f.create_dataset('train/features_categorical', data=train_categorical_encoded.to_numpy()[train_indices])
    f.create_dataset('train/features_numerical', data=train_numerical.to_numpy()[train_indices])
    f.create_dataset('train/click', data=train_click.to_numpy()[train_indices])
    f.create_dataset('train/conversion', data=train_conversion.to_numpy()[train_indices])

    f.create_dataset('val/features_categorical', data=train_categorical_encoded.to_numpy()[val_indices])
    f.create_dataset('val/features_numerical', data=train_numerical.to_numpy()[val_indices])
    f.create_dataset('val/click', data=train_click.to_numpy()[val_indices])
    f.create_dataset('val/conversion', data=train_conversion.to_numpy()[val_indices])

    f.create_dataset('test/features_categorical', data=test_categorical_encoded.to_numpy())
    f.create_dataset('test/features_numerical', data=test_numerical.to_numpy())
    f.create_dataset('test/click', data=test_click.to_numpy())
    f.create_dataset('test/conversion', data=test_conversion.to_numpy())
```
